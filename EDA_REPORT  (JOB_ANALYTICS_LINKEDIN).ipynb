#  Exploratory Data Analysis (E.D.A.)

## 2.1 Job Openings As Per State

```python
plt.figure(figsize=(13,6), dpi=150)
chart1 = sns.countplot(df["Location"],palette='mako')
plt.title("Job Openings by State",size=25)
plt.ylabel("Number of job openings",size=20)
plt.xlabel("State",size=20)
plt.xticks(rotation=80)
chart1.bar_label(chart1.containers[0],size=12)
```

## 2.2 Jobs Available In Each Industry.

```python
plt.figure(figsize=(13,6), dpi=150)
chart2 = sns.countplot(df["Industry"],palette='twilight')
plt.title("Job Openings by Industry",size=25)
plt.ylabel("Number of job openings",size=20)
plt.xlabel("Industry",size=20)
plt.xticks(rotation=30,fontsize=12)
chart2.bar_label(chart2.containers[0],size=12)
```

## 2.3 Jobs Available In  Company As Per Employee Count

```python
plt.figure(figsize=(13,6), dpi=150)
chart3 = sns.countplot(df["Company_Name"],hue=df["Employee_count"],palette='Blues_r')
# plt.title("Job Openings by Company",size=25)
plt.ylabel("Number of job openings",size=20)
plt.xlabel("Company",size=20)
plt.xticks(rotation=70)
chart3.bar_label(chart3.containers[3],size=13)
```

## 2.4 Job Openings At Each Level

```python
plt.figure(figsize=(13,6), dpi=200)
chart=sns.countplot(df["Level"],palette='ocean')
plt.title("Job Openings at different levels",size=25)
plt.ylabel("Number of job openings",size=20)
plt.xlabel("Level",size=20)
chart.bar_label(chart.containers[0],size=13)
```

## 2.5  Job Openings For Different Involvement Nature

```python
plt.figure(figsize=(13,6), dpi=200)
chart = sns.countplot(df["Involvement"],palette='ocean_r')
plt.title("Job Openings for different involvement natures",size=25)
plt.ylabel("Number of job openings",size=20)
plt.xlabel("Involvement nature",size=20)
chart.bar_label(chart.containers[0],size=13)
```

## 2.6 Checking Relation Between Linkedin Follower And Employee Count 

```python
plt.figure(figsize=(20,10),dpi=100)
chart = sns.scatterplot(data=df,x="LinkedIn_Followers",y="Employee_count",hue="Company_Name",s=400,palette='Dark2')
plt.title("Relation between linkedin follower & employee count for each company",size=25)
plt.ylabel("Employee Count",size=22)
plt.xlabel("Followers",size=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
chart.bar_label(chart.containers[0],size=13)
```

## 2.10  Jobs that require Specific particular skill

```python
df_skills=df.drop(["Company_Name","Class","Designation","Location","Total_applicants","LinkedIn_Followers","Involvement","Involvement","Level","Employee_count","Industry"],axis=1)
skill_dic={}
for i in df_skills.columns:
    skill_dic[i]=df[i].sum()
data = list(skill_dic.items())
skill_df = pd.DataFrame(data, columns=['Skill', 'Number of jobs requiring the skill'])
```

```python
skill_df
```

```python
plt.figure(figsize=(20,10),dpi=100)
chart = sns.barplot(x=skill_df["Skill"],y=skill_df["Number of jobs requiring the skill"],palette="icefire")
plt.title("Number of jobs that require a particular skill",size=25)
plt.ylabel("Number of jobs",size=22)
plt.xlabel("Skill",size=22)
plt.xticks(fontsize=18,rotation=90)
plt.yticks(fontsize=18)
chart.bar_label(chart.containers[0],size=18)
```
